```python
# Loading csv path data into Pandas using read_csv

crypto_tmc_path = Path("./crypto_tmc.csv")
crypto_tmc_data = pd.read_csv(crypto_tmc_path, index_col="date", infer_datetime_format=True, parse_dates=True)
```
